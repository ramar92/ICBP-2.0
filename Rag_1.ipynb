```python
from datasets import load_dataset

# Load a sample Wikipedia dataset (or replace with your own)
dataset = load_dataset("wikimedia/wikipedia", "20231101.en", split="train[:1000]")
docs = [doc["text"] for doc in dataset if doc["text"].strip() != ""]
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

embedder = SentenceTransformer("all-MiniLM-L6-v2")
doc_embeddings = embedder.encode(docs, convert_to_numpy=True)
```

```python
import faiss

index = faiss.IndexFlatL2(doc_embeddings.shape[1])
index.add(doc_embeddings)
```

```python
def retrieve_docs(query, k=5):
    query_embedding = embedder.encode([query])
    distances, indices = index.search(query_embedding, k)
    return [docs[i] for i in indices[0]]
```

```python
from transformers import pipeline

# Use a summarization-style model like BART for generation
generator = pipeline("text2text-generation", model="facebook/bart-large")
```

```python
def rag_generate(query):
    retrieved = retrieve_docs(query, k=3)
    # Limit context length to prevent exceeding model's max sequence length
    context = "\n".join(retrieved)[:512]
    prompt = f"Context:\n{context}\n\nQuestion: {query}\nAnswer:"
    answer = generator(prompt, max_new_tokens=100)[0]["generated_text"]
    return answer

# Try it!
query = "What are the health benefits of green tea?"
print(rag_generate(query))
```
